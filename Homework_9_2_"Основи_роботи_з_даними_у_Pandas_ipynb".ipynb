Відповідь напишіть текстом в окремій клітинці типу Markdown.

Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
yangon_mean = df[df['City'] == 'Yangon']['gross income'].mean()
mandalay_mean = df[df['City'] == 'Mandalay']['gross income'].mean()

print("Середнє gross income у Yangon:", yangon_mean)
print("Середнє gross income у Mandalay:", mandalay_mean)

if yangon_mean > mandalay_mean:
    print("У Yangon середній gross income більший.")
elif mandalay_mean > yangon_mean:
    print("У Mandalay середній gross income більший.")
else:
    print("Середні gross income у Yangon та Mandalay однакові.")
```

8. Для продовження дослідження різниці у прибутках в магазинах давайте зʼясуємо, чи відрізняється маржа ('gross margin percentage') для різних інвойсів в магазині? Перевірте це програмно на даних виходячи з того функціоналу, який ми вже пройшли у лекціях.

Дайте на основі отриманого результату відповідь: "У всіх всіх інвойсах однакова маржа" або "Маржа відрізняється для різних інвойсів".

```python
# Групуємо дані по інвойсу та дивимось унікальні значення маржі
margin_per_invoice = df.groupby('Invoice ID')['gross margin percentage'].nunique()

# Перевіряємо, чи є інвойси з різною маржею (більше ніж 1 унікальне значення)
if (margin_per_invoice > 1).any():
    print("Маржа відрізняється для різних інвойсів")
else:
    print("У всіх інвойсах однакова маржа")
```

9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
# Відфільтруємо дані для Yangon і Mandalay
yangon_df = df[df['City'] == 'Yangon']
mandalay_df = df[df['City'] == 'Mandalay']

# Порахувати кількість унікальних інвойсів по категоріях для Yangon
yangon_invoice_counts = yangon_df.groupby('Product line')['Invoice ID'].nunique()
yangon_invoice_percentage = (yangon_invoice_counts / yangon_invoice_counts.sum())

# Порахувати кількість унікальних інвойсів по категоріях для Mandalay
mandalay_invoice_counts = mandalay_df.groupby('Product line')['Invoice ID'].nunique()
mandalay_invoice_percentage = (mandalay_invoice_counts / mandalay_invoice_counts.sum())

print("Відсотковий розподіл кількості інвойсів по категоріях товарів у Yangon:")
print(yangon_invoice_percentage.sort_values(ascending=False))

print("\nВідсотковий розподіл кількості інвойсів по категоріях товарів у Mandalay:")
print(mandalay_invoice_percentage.sort_values(ascending=False))
```
